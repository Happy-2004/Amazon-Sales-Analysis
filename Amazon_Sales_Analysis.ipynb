```python
# Use the .str.replace() command to remove commas and ₹ from the discounted_price and actual_price columns.
# Use the .astype() command to convert the data type in the discounted_price and actual_price columns to float64.
df['discounted_price'] = df['discounted_price'].str.replace(',','').str.replace('₹','').astype('float64')
df['actual_price'] = df['actual_price'].str.replace(',','').str.replace('₹','').astype('float64')
```

```python
# Creating the discount_amount column.

df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
# Converting the discount_percentage column.

df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# Creating a column that ranks products based on rating score.

ranking = []

for score in df['rating']:
  if score <= 0.9 : ranking.append('Very Poor')
  elif score <= 1.9 : ranking.append('Poor')
  elif score <= 2.9 : ranking.append('Average')
  elif score <= 3.9 : ranking.append('Good')
  elif score <= 4.9 : ranking.append('Very Good')
  elif score <= 5.0 : ranking.append('Excellent')

df['ranking'] = ranking

df['ranking'] = df['ranking'].astype('category')
```

### **Create a dataframe with reviewer data**

```python
# Use the .str.split() command to split string values in the user_id and user_name columns using , as a delimiter.
# expand = False will insert every split string element into a list.

df['user_id'] = df['user_id'].str.split(',', expand=False)
df['user_name'] = df['user_name'].str.split(',', expand=False)
```

### **Linear Regression Analysis**

```python
# creating correlation matrix.

df_correl = df_clean[['actual_price','discounted_price','discount_percentage','rating','rating_count']]
df_correl
```

```python
# creating heatmap for visualizing reltionship between two variables.

sns.heatmap(df_correl.corr(), annot=True, vmax=1, vmin=-1, center=0, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# actual_price and discounted_price has the good positive relationship.
# plotting scatter plot between them.

sns.scatterplot(x=df_correl['actual_price'], y=df_correl['discounted_price'])
plt.title('Actual Price vs Discounted Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
# A linear regression model can be created to predict the discounted price of a product based on the product's actual price.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = df_correl['actual_price'].values.reshape(-1,1)
y = df_correl['discounted_price'].values.reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Actual Price vs Discounted Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
# checking r2 score of the model.

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(y_test, y_predict)
print('R2 = ', r2)

mse = mean_squared_error(y_test, y_predict)
print('MSE = ', mse)

mae = mean_absolute_error(y_test, y_predict)
print('MAE = ', mae)
```

```python
plt.scatter(x=x_test, y=y_test)
plt.plot(x_test, model.predict(x_test), color='red')
plt.title('Actual Price vs Discounted Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```
